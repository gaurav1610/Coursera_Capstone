# Importing Libraries 

```python
# import wikipedia
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
```

URL 

```python
# first pull the HTML from the page that links to all of the pages with the links.
# in this case, this page gives the links list pages of sci-fi films by decade.
# just go to https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:M
# to see what I'm pulling from
html = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

#turn the HTML into a beautiful soup text object
b = BeautifulSoup(html.text, 'lxml')
```

```python
#print(b.prettify())
```

```python
My_table = b.find('table',{'class':'wikitable sortable'})
My_table
```

# TABLE SCRAPING

```python
Postcode = []
Borough = []
Neighbourhood = []
for row in My_table.find_all('tr')[1:]:
    Borough_cell = row.find_all('td')[1]
    if Borough_cell.text!='Not assigned':
      Postcode_cell = row.find_all('td')[0]
      Postcode.append(Postcode_cell.text)
      Borough.append(Borough_cell.text)
      Neighbourhood_cell = row.find_all('td')[2]
      neighbourhood=Neighbourhood_cell.text.rstrip('\n')
      if neighbourhood == 'Not assigned': neighbourhood=Borough_cell.text
      Neighbourhood.append(neighbourhood)
```

# CONVERTING INTO DATAFRAME

```python
import pandas as pd
df=pd.DataFrame()
df['Postcode'] = Postcode
df['Borough'] = Borough
df['Neighbourhood'] = Neighbourhood
df
```
